```python
# List to hold results
results = []

for url in urls:
    try:
        response = requests.get(url)
        response.raise_for_status()  # Check for HTTP errors
        data = response.json()
        
        # Extract D and dD values
        mag_field = data.get('magneticFields', {})
        D_value = mag_field.get('D', None)
        dD_value = mag_field.get('dD', None)
        
        # Parse strings with units if necessary
        def parse_value(val):
            if isinstance(val, str):
                # Extract numeric part from string, e.g., "3.969 deg"
                match = re.search(r"[-+]?\d*\.\d+|\d+", val)
                if match:
                    return float(match.group())
            elif isinstance(val, (int, float)):
                return val
            return None
        
        D_numeric = parse_value(D_value)
        dD_numeric = parse_value(dD_value)
        
        # Store the extracted info
        results.append({
            'url': url,
            'D': D_numeric,
            'dD': dD_numeric
        })

    except requests.RequestException as e:
        print(f"Error fetching {url}: {e}")
    except ValueError:
        print(f"Invalid JSON response from {url}")

# Export to CSV without units
with open('station_declination_data.csv', 'w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=['D', 'dD'])
    writer.writeheader()
    for row in results:
        # Write numeric values directly
        writer.writerow({key: row[key] for key in ['D', 'dD']})

print("Data exported to 'station_declination_data.csv'")
```
